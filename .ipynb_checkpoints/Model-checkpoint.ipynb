## Utils

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix , ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, target_names=le.classes_)
    print("\nClassification Report:\n")
    print(report)
    cm = confusion_matrix(y_test, y_pred)
    cm_df = pd.DataFrame(cm, index=le.classes_, columns=le.classes_)
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.show()
```

```python
def evaluate_deep_model(model, X_test_tensor, y_test_tensor):
    torch.cuda.empty_cache()
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
    model.eval()  # Set model to eval mode
    
    all_preds = []
    with torch.no_grad():
        for inputs, _ in test_loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, dim=1)
            all_preds.append(preds.cpu())
    y_pred = torch.cat(all_preds).numpy()
    y_true = y_test_tensor.cpu().numpy()

    # Classification report
    report = classification_report(y_true, y_pred, target_names=le.classes_)
    print("\nClassification Report:\n")
    print(report)

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    cm_df = pd.DataFrame(cm, index=le.classes_, columns=le.classes_)
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.show()
```

```python
def evaluate_autoencoder(model, X_test_tensor, y_test_tensor, threshold=None, le=None):
    model.eval()  # evaluation mode
    test_loader = torch.utils.data.DataLoader(X_test_tensor, batch_size=64, shuffle=False)
    criterion = torch.nn.MSELoss(reduction='none')  # per-sample reconstruction loss

    reconstruction_errors = []

    with torch.no_grad():
        for inputs in test_loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            # Compute MSE per sample (average feature-wise error)
            loss = torch.mean(criterion(outputs, inputs), dim=1)
            reconstruction_errors.extend(loss.cpu().numpy())

    reconstruction_errors = np.array(reconstruction_errors)

    # If threshold not provided, compute dynamically (95th percentile of benign scores)
    if threshold is None:
        threshold = np.percentile(reconstruction_errors, 95)
        print(f"Auto-calculated anomaly threshold: {threshold:.6f}")

    # Classify anomalies based on threshold
    y_pred = (reconstruction_errors > threshold).astype(int)
    y_true = y_test_tensor.cpu().numpy().astype(int)

    # Classification metrics
    print("\nClassification Report (0: Benign, 1: Attack):\n")
    print(classification_report(y_true, y_pred, target_names=['Benign', 'Attack']))

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    labels = ['Benign', 'Attack']
    cm_df = pd.DataFrame(cm, index=labels, columns=labels)

    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Anomaly Detection Confusion Matrix')
    plt.tight_layout()
    plt.show()

    # Reconstruction error distribution
    plt.figure(figsize=(8,4))
    sns.histplot(reconstruction_errors[y_true==0], bins=50, label="Benign", color='green', alpha=0.6)
    sns.histplot(reconstruction_errors[y_true==1], bins=50, label="Attack", color='red', alpha=0.6)
    plt.axvline(threshold, color='black', linestyle='--', label='Threshold')
    plt.xlabel('Reconstruction Error')
    plt.ylabel('Frequency')
    plt.legend()
    plt.title('Reconstruction Error Distribution')
    plt.show()
```

### Model : ResNet like Classifier

#### Residual Block

```python
import torch.nn as nn
import torch.nn.functional as F

class ResidualBlock(nn.Module):
    def __init__(self, dim, dropout=0.2):
        super(ResidualBlock, self).__init__()
        self.lin1 = nn.Linear(dim,dim)
        self.bn1 = nn.BatchNorm1d(dim)
        self.lin2 = nn.Linear(dim,dim)
        self.bn2 = nn.BatchNorm1d(dim,dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        residual = x
        out = F.relu(self.bn1(self.lin1(x)))
        out = self.dropout(out)
        out = self.bn2(self.lin2(out))
        out += residual
        out = F.relu(out)
        return out
```

```python
class ResNetClassifier(nn.Module):
    def __init__(self, in_dim, hidden_dim, num_blocks, num_classes, dropout = 0.2):
        super(ResNetClassifier, self).__init__()
        self.input_layer = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU()
        )

        self.residual_blocks = nn.Sequential(
            *[ResidualBlock(hidden_dim, dropout) for _ in range(num_blocks)]
        )
        self.dropout = nn.Dropout(dropout)
        self.output_layer = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        out = self.input_layer(x)
        out = self.residual_blocks(out)
        out = self.dropout(out)
        out = self.output_layer(out)
        return out
```

```python
input_dim = X_train.shape[1]
num_classes = len(le.classes_)

model = ResNetClassifier(input_dim, hidden_dim=256, num_blocks=4, num_classes=num_classes)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model
```

### Training Loop

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs = 20

for epoch in range(num_epochs):
    model.train()    # set model to training mode
    running_loss = 0.0
    correct = 0
    total = 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()  # Zero gradients before backward pass
        outputs = model(inputs)  # Forward pass

        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()*inputs.size(0)
        _, predicted = torch.max(outputs, dim=1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    epoch_loss = running_loss / total
    epoch_acc = correct / total

    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {epoch_loss:.4f} - Accuracy: {epoch_acc:.4f}")
    torch.save(model.state_dict(), 'model_weights.pth')
```

## Encoder-Decoder : Anomaly Detection

### Model

```python
class Autoencoder(nn.Module):
    def __init__(self, input_dim=78,dropout=0.2):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(dropout),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(dropout),

            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),

            nn.Linear(32, 16),
            nn.ReLU()
        )

        self.decoder = nn.Sequential(
            nn.Linear(16, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),

            nn.Linear(32, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),

            nn.Linear(64, 128),
            nn.ReLU(),

            nn.Linear(128, 256),
            nn.ReLU(),

            nn.Linear(256, input_dim)  
        )

    
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

### Creating new train-test for anomaly detection
- train-set = Benign examples
- test-set = Attacks + Benign Examples

```python
data_benign = data[data['Label'] == 0]
data_benign.shape
```

```python
data_attack = data[data['Label'] != 0]
data_attack.shape
```

```python
train_size = int(0.1 * data_benign.shape[0])  # 10% of benign data
ad_X_train = data_benign[:-train_size].copy()        # 90% benign for training
ad_X_test = pd.concat([data_benign[-train_size:].copy(), data_attack], ignore_index=True)
```

```python
ad_X_train = ad_X_train.drop(['Label'], axis=1)
ad_y_test = ad_X_test['Label'].apply(lambda x : 0 if x == 0 else 1)
ad_X_test = ad_X_test.drop(['Label'], axis=1)
```

```python
# standardization
ad_std = StandardScaler()
ad_X_train_scaled = ad_std.fit_transform(ad_X_train)
ad_X_test_scaled = ad_std.transform(ad_X_test)
```

```python
ad_X_train_tensor = torch.tensor(ad_X_train_scaled, dtype=torch.float32)
ad_X_test_tensor = torch.tensor(ad_X_test_scaled, dtype=torch.float32)
ad_y_test_tensor = torch.tensor(ad_y_test.values,dtype = torch.long)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

ad_X_train_tensor = ad_X_train_tensor.to(device)
ad_X_test_tensor = ad_X_test_tensor.to(device)
ad_y_test_tensor = ad_y_test_tensor.to(device)

ad_train_dataset = TensorDataset(ad_X_train_tensor)
ad_train_loader = DataLoader(ad_train_dataset, batch_size=256, shuffle=True)
```

```python
import torch.nn as nn
import torch.optim as optim
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = Autoencoder(input_dim=ad_X_train.shape[1]).to(device)

criterion = nn.MSELoss()  # reconstruction loss
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)  # L2 regularization
```

```python
# Early stopping parameters
patience = 5
best_loss = np.inf
trigger_times = 0

num_epochs = 100

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for (inputs,) in ad_train_loader:
        inputs = inputs.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {epoch_loss:.6f}")

    # Early stopping check
    if epoch_loss < best_loss:
        best_loss = epoch_loss
        trigger_times = 0
        torch.save(model.state_dict(), 'best_autoencoder.pth')
        print(f"Loss improved. Model saved.")
    else:
        trigger_times += 1
        print(f"No improvement. Trigger times: {trigger_times}/{patience}")
        if trigger_times >= patience:
            print("Early stopping triggered.")
            break
```

```python
model = Autoencoder(input_dim=78)
model.load_state_dict(torch.load('best_autoencoder.pth'))
model.to(device)
```

```python
import re

# Create mapping: original → cleaned name
def clean_and_map_feature_names(df):
    clean_names = []
    name_map = {}
    
    for col in df.columns:
        # Replace all non-alphanumeric characters with underscores
        clean_col = re.sub(r'[^A-Za-z0-9_]+', '_', col)
        
        # If duplicate after cleaning, make it unique
        if clean_col in clean_names:
            i = 1
            while f"{clean_col}_{i}" in clean_names:
                i += 1
            clean_col = f"{clean_col}_{i}"
        
        clean_names.append(clean_col)
        name_map[clean_col] = col  # Map cleaned → original
    
    df = df.copy()
    df.columns = clean_names
    return df, name_map


# Example usage:
X_train_clean, feature_map = clean_and_map_feature_names(X_train)
X_test_clean = X_test.copy()
X_test_clean.columns = X_train_clean.columns  # ensure same names
```
